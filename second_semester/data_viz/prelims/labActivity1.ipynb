```python
import pandas as pd
import matplotlib.pyplot as plt
import os 
```

```python
data_frame = {
            'Quarter': ['Q1 2025', 'Q2 2025', 'Q3 2025', 'Q4 2025', 'Q1 2026'],
            'Net Revenue ($M)': [420.0, 435.5, 448.2, 470.4, 482.0],
            'Operating Cost ($M)': [360.0, 372.0, 378.0, 385.0, 390.0],
            'Net Profit ($M)': [60.0, 63.5, 70.2, 85.4, 92.0],
            'Growth vs. PY':  [8.2, 9.1, 10.5, 12.0, 14.7],
}
```

```python
df = pd.DataFrame(data_frame)
df
```

```python
plt.plot(df['Quarter'], df['Net Revenue ($M)'], marker='o', label='Net Revenue ($M)')
plt.plot(df['Quarter'], df['Operating Cost ($M)'], marker='s', label='Operating Cost ($M)')
plt.xlabel('Quarter')
plt.ylabel('Amount ($M)')
plt.title('Financial Performance Over Time')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.bar(df['Quarter'], df['Growth vs. PY'], color='skyblue')
plt.xlabel('Quarter')
plt.ylabel('Growth vs. PY (%)')
plt.title('Quarterly Growth vs. Previous Year')
plt.grid(True)
plt.savefig("my_plot.pdf", bbox_inches='tight') # 'bbox_inches="tight"' removes extra whitespace
plt.show()
```
