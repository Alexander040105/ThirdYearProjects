```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import mlxtend
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
current_dir = os.getcwd()
df = pd.read_csv(f'{current_dir}\\strong_association_transactions.csv')
df_copy = df.copy()
df.isna().sum()
```

```python
df_copy
```

```python
transactions = []
for i in range(len(df_copy)):
    df_copy.iloc[i] = df_copy.iloc[i].str.replace(',', '')
    for j in df_copy.iloc[i]:
        if j != 'nan' and j != '':
            transactions.append(str(j).split())
transactions
```

```python
te = TransactionEncoder()
df_copy = te.fit_transform(transactions)

df_copy
```

```python
df_copy = pd.DataFrame(df_copy, columns=te.columns_)
df_copy
```

```python
frequent_itemsets = apriori(df_copy, min_support=0.4, use_colnames=True)
frequent_itemsets
```

```python
res = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.4)
results = res[(res['confidence'] > 0.6) & (res['lift'] > 1.2)][['antecedents', 'consequents', 'support', 'confidence', 'lift']].head(5).sort_values(by='confidence', ascending=False)
```

```python
results[['support','confidence','lift']].apply(np.mean)
```
