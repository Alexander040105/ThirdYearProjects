```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import mlxtend
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
current_dir = os.getcwd()
df = pd.read_csv(f'{current_dir}\\Market_Basket_Optimisation.csv')
df_copy = df.copy()
df.isna().sum()
```

```python
transactions = []
for i in range(len(df)):
    transaction = [str(df.values[i, j]) for j in range(20) if str(df.values[i, j]) != 'nan']
    transactions.append(transaction)
```

```python
# df_copy = pd.get_dummies(df_copy, columns=['itemDescription'])

te = TransactionEncoder()
df_copy = te.fit_transform(transactions)

df_copy
```

```python
df_copy = pd.DataFrame(df_copy, columns=te.columns_)
df_copy
```

```python
frequent_itemsets = apriori(df_copy, min_support=0.009, use_colnames=True)
frequent_itemsets
```

```python
res_confidence = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.3)
res_confidence
```

```python
res_lift = association_rules(frequent_itemsets, metric="lift", min_threshold=2.5)
res_lift
```

```python
res_leverage = association_rules(frequent_itemsets, metric="leverage", min_threshold=0.02)
res_leverage
```

```python
res_support = association_rules(frequent_itemsets, metric="support", min_threshold=0.009)
res_support
```

```python
res_certainty = association_rules(frequent_itemsets, metric="certainty", min_threshold=0.3)
res_certainty
```
