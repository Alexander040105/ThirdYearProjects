```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import mlxtend
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
current_dir = os.getcwd()
df = pd.read_csv(f'{current_dir}\\Groceries_dataset.csv')
df_copy = df.copy()
df.isna().sum()
```

```python
df['itemDescription'].unique()
```

```python
df['itemDescription'].value_counts()
```

```python
# df_copy = pd.get_dummies(df_copy, columns=['itemDescription'])
df_copy.drop(columns=['Member_number', 'Date'], inplace=True)

te = TransactionEncoder()
df_copy = te.fit_transform(df_copy['itemDescription'].apply(lambda x: [x]))

df_copy
```

```python
df_copy = pd.DataFrame(df_copy, columns=te.columns_)
df_copy
```

```python
frequent_itemsets = apriori(df_copy, min_support=0.01, use_colnames=True)
frequent_itemsets
```

```python
res_confidence = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.6)
res_confidence
```
